## Exercise 4 — Medical Screening Simulation

```python
import numpy as np

np.random.seed(42)

n = 10000
prevalence = 0.01
sensitivity = 0.99
specificity = 0.95

disease = np.random.binomial(1, prevalence, n)

test = []

for d in disease:
    if d == 1:
        test.append(np.random.binomial(1, sensitivity))
    else:
        test.append(np.random.binomial(1, 1-specificity))

screen = pd.DataFrame({'disease': disease, 'test_result': test})
screen.head()
```

## Exercise 5 — Confusion Matrix for Screening Test

```python
from sklearn.metrics import confusion_matrix

confusion_matrix(screen['disease'], screen['test_result'])
```

## Exercise 6 — Bayes Interpretation

```python
true_positive = screen[(screen.disease==1) & (screen.test_result==1)].shape[0]
false_positive = screen[(screen.disease==0) & (screen.test_result==1)].shape[0]

posterior = true_positive/(true_positive+false_positive)
posterior
```

## Exercise 9 — Logistic Regression Model

```python
import statsmodels.api as sm

X = df[['Glucose','BMI','Age']]
y = df['Outcome']

X = sm.add_constant(X)

model = sm.Logit(y, X).fit()
print(model.summary())
```

## Exercise 10 — Model Evaluation

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix

X = df.drop('Outcome', axis=1)
y = df['Outcome']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

pred = model.predict(X_test)

print('Confusion Matrix:')
print(confusion_matrix(y_test, pred))

print('Accuracy:', accuracy_score(y_test, pred))
```
